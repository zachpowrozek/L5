# 3. Creating Strings with Quotes

### Different Ways Quotes Work in Strings

Python lets you create strings with **single quotes**, **double quotes**, or **triple quotes**.

- **Single quotes**: `'Inception'`
- **Double quotes**: `"Inception"`
- **Triple quotes**: `'''Inception'''` or `"""Inception"""` (often used for multi-line text)

These are all valid and create the same string:
```python
title1 = 'Inception'
title2 = "Inception"
title3 = """Inception"""
```

Use whichever is more convenient.

If your string contains a single quote ('), wrap the string in double quotes.

If it contains a double quote ("), wrap it in single quotes.

If it contains both, you can either escape them with a backslash (\' or \") or use triple quotes.

```
quote = 'He said, "I see dead people."'
```

````python
# Demo: Strings can be created with different quote styles.
# Goal: Practice storing and printing a string variable.

movie1 = '“Who’s Afraid of Virginia Woolf?”'
print(f"The first movie is {movie1}.")
````

# 5. Indexing & Slicing Strings

Slicing lets you take out a *range of characters* from a string.  
The general form is:
```
string[start:end]
```

- **start** = the index where the slice begins (inclusive).  
- **end** = the index where the slice ends (exclusive).  
- If you leave out **start**, it defaults to the beginning.  
- If you leave out **end**, it goes to the end.  
- You can also use **negative indices** to count from the right side.  

Examples (using `"Moonlight"`):  
- `title[0:4]` → `"Moon"` (characters at positions 0, 1, 2, 3).  
- `title[:4]` → `"Moon"` (same as above, since start defaults to 0).  
- `title[4:]` → `"light"` (from index 4 through the end).  
- `title[-3:]` → `"ght"` (the last 3 characters).  


Important: slicing creates a **new string**. The original string does not change.

This means that you should save the result in another variable.  


````python
# Demo: Indexing and slicing a string.
# Goal: See how to select a single character and a range of characters.

title = "Moonlight"
print(f"The first letter of {title} is {title[0]}.")
print(f"The slice title[:4] of {title} gives {title[:4]}.")
print(f"The slice title[4:8] of {title} gives {title[4:8]}.")
````

# 7. String Methods: Acting on Text

String **methods** are special functions that belong to string objects.  
The pattern always looks like this:
```
string.method(optional_arguments)
```

- **string** → can be a hardcoded string (`"Inception"`) or a variable (`title`).  
- **. (dot)** → tells Python “use a method on this object.”  
- **method name** → the action you want to perform (like `upper`, `strip`, or `find`).  
- **optional arguments** → extra information the method might need (for example, what text to search for).

Examples:
```python
"coco".upper()
# 'COCO'

title = "   Inception   "
title.strip()
# 'Inception'
```



# 8. Saving the Results of String Methods

## Remember! String Methods ALWAYS Return New Values
Important: Strings are immutable, so these methods always return a new string or value. The original variable does not change.


This is something students often forget to do -- they print it, but don't save it anywhere.

````python
# Demo: String methods return new values (they usually do not change the original string).
# Goal: Notice that calling a method by itself may not display output unless printed.

"coco".upper()
# 'COCO'

title = "   Inception   "
title.strip()
# 'Inception'

print("Did you get any output?  How can we save the values we created?")
````

# 11. Boolean String Methods

These methods return **True** or **False** depending on a condition.  
Examples include:
- `.isupper()` → True if all characters are uppercase  
- `.islower()` → True if all characters are lowercase  
- `.startswith("sub")` → True if string starts with `"sub"`  
- `.endswith("sub")` → True if string ends with `"sub"`  
- `.isalnum()` → True if string has only letters and numbers  



Note: You wouldn’t just use `"in"` because these methods are more specific.  
For example:
```python
"Avatar".startswith("Av")   # True
"Avatar".endswith("tar")    # True
"A" in "Avatar"             # True, but not as precise

Use `in` for “contains anywhere.”
Use `.startswith()` or `.endswith()` when you want a specific position.


````python
# Demo: Boolean string methods return True/False.
# Goal: Test properties of a string (prefix check, uppercase check).

movie3 = "Inception"
print(f"Does {movie3} start with 'in'? {movie3.startswith('in')}")
print(f"Is {movie3} all uppercase? {movie3.isupper()}")
````

# 13. Searching and Counting in Strings

These methods return a **number** that tells you something about the string.  

Common examples:
- `.find("sub")` → returns the index of the first occurrence of `"sub"`, or `-1` if not found  
- `.index("sub")` → like `find`, but gives an **error** if not found  
- `.count("sub")` → counts how many times a substring appears  

Examples:
```python
title = "Interstellar"
print(title.find("l"))      # 5 (first 'l')
print(title.find("l", 6))   # 9 (looks later in the string)

print(title.count("l"))     # 3 (there are 3 lowercase l's)
print(title.count("L"))     # 0 (uppercase L is different)


Note: String searches are case-sensitive.

````python
# Demo: `.find()` and `.count()` return integers.
# Goal: See how to search for a character and count occurrences (case-sensitive).

title = "Interstellar"
print(title.find("z"))        # position of "z" (returns -1 if not found)
print(title.find("l", 6))     # find "l" starting from index 6
print(title.count("l"))       # count how many lowercase "l"
print(title.count("L"))       # count how many uppercase "L" (case-sensitive)
````

# 17. Counting in Strings vs. Lists

## Remember!: even though some operations look the same, they aren't.

The `.count()` method with a string returns partial matches but when used on a list it only returns exact matches.  

- For strings, `.count("a")` counts characters inside the text.

- For lists, `.count("a")` counts exact list elements.

They look similar, but behave differently.


Examples:
```python
title = "Coco"
print(len(title))          # 4
print("o" in title)        # True
print(title.count("o"))    # 2

genres = ["Sci-Fi", "Action", "Drama"]
print(len(genres))         # 3
print("Action" in genres)  # True
print(genres.count("Drama"))  # 1

```

````python
# Demo: `.count()` behaves differently for strings vs. lists.
# Goal: Count characters in a string vs. exact elements in a list.

str_variable = "The cat in the hat"            # a single string
list_variable = str_variable.split()           # a list of words

print(str_variable)                            # print the original string
print(list_variable)                           # print the word list

print(f"The letter `a` appears {str_variable.count('a')} times in {str_variable}")  # count character 'a'
print(f"The word `a` appears {list_variable.count('a')} times in {list_variable}")  # count exact word "a"
````
